```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/4.csv', index_col=0)

# Create a line chart for each country
fig, ax = plt.subplots(figsize=(10, 6))
for country in df.columns:
    ax.plot(df.index, df[country], label=country)

# Set the chart title and labels
ax.set_title('Inland Fish Production by USA')
ax.set_xlabel('Year')
ax.set_ylabel('Production (tonnes)')

# Add a legend
ax.legend()

# Show the chart
plt.show()
```

```python
df.describe()
```

```python
x=df[['2009','2003','2004','2005','2006','2007','2008','2009']]
y=df[['2010','2011']]
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.25)
```

```python
reg = LinearRegression()
model = reg.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import math

print('Mean absolute error:', mean_absolute_error(y_test, y_pred))

print("Mean squared error:", mean_squared_error(y_test, y_pred))
print('variance score:',r2_score(y_test, y_pred))

print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, y_pred)))
```
